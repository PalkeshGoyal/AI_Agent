```python
from langchain.agents import create_agent
from langchain_classic.agents import AgentExecutor , create_react_agent, create_openai_tools_agent, create_tool_calling_agent
from langchain_classic import hub
```

```python
# prompt = hub.pull("hwchase17/react")
prompt = hub.pull("hwchase17/openai-tools-agent")
```

```python
# prompt.template
```

```python
llm = ChatOpenAI(model="gpt-5-nano")
```

```python
agent = create_tool_calling_agent(
    llm=llm,
    tools=[search_tool],
    prompt=prompt
    )
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool],
    verbose=True
)
```

```python
response = agent_executor.invoke(
    {
        "input" : "Recent indian news"
    }
    ,
    verbose=True
)
print(response)
```

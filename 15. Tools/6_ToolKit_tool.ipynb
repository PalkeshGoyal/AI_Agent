```python
from langchain.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

@tool   
def add(a: int, b: int) -> int:
    """Add two numbers."""
    return a + b
```

```python
class MathToolkit:
    """A toolkit for math operations."""

    tools = [multiply, add] 

    def get_tools(self):
        return self.tools
```

```python
toolkit = MathToolkit()
tool_list = toolkit.get_tools()
for t in tool_list:
    print(t.name, ":", t.description)
```

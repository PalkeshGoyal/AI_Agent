```python
from langchain.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b
```

```python
result = multiply.invoke({"a": 6, "b": 7})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
# print(multiply.return_direct)
print(multiply.args)
```
